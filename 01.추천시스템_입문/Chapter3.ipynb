# CH 3 추천 시스템의 UI/UX
- 추천 시스템이 기능하기 위해 제공해야 할 UI/UX에 관해 소개
- 추천 결과를 전달할 때 어떤 화면(UI)에 따라 어떤 체험(UX)을 제공해야 하는지에 주목
- 이 부분은 1장의 추천 시스템 3가지 구성 요소 중 출력(추천 결과 제시)에 해당

## 3.1 UI/UX의 중요성
- 추천 알고리즘으로 선택된 아이템이 최적의 형태로 전달되지 못하면 가치도 전달되지 않음
- 추천 시스템을 활용하는 서비스의 특징적인 UI/UX에 관해 소개

## 3.2 서비스를 사용하는 사용자의 목적에 적합한 UI/UX 사례
- 사용자 목적별로 UI/UX의 사례 소개
- j.Herlocker를 참고해서 4가지 분류
  - 적합 아이템 발견
  - 적합 아이템 나열
  - 아이템 계열 소비
  - 서비스 내 체류

### 3.2.1 적합 아이템 발견
- 사용자가 자신의 목적을 달성하는 데 적합한 아이템을 서비스상에서 최소 1개 이상 발견하는 경우
- 사용자 기호에 맞을 가능성이 높은 아이템부터 차례대로 정렬한 리스트를 사용자에게 제시하는 것이 효과적
  - 리스트 형식으로 표시하는 것은 아이템 하나당 표시할 수 있는 정보량이 제한됨 -> 정보를 취사선택해 리스트에 표시해야 함
- 아래는 몇가지 대안
  - 표준, 랭킹, 리뷰 수, 신규 오픈 등의 리스트로 표현
  - 넷틀릭스의 경우 스와이프로 화면 이동 없이 많은 리스트를 표현
  - 틴더는 아이템(사람)을 좌우로 슬라이드하여 사용자로부터 선호 여부를 즉각 피드백
  - 긍정, 부정 비드백을 활용한 추천 시스템 구축을 고려할 수 있음

### 3.2.2 적합 아이템 나열
- 사용자가 자신의 목적을 달성하는 데 적합한 아이템을 가능한 모든 서비스에서 발견하는 경우
- 사용자가 여러 아이템을 열람할 필요가 있는 적합 아이템 나열에서는 일관성 있는 표시 형식으로 제공하는 것이 일반적
- 얼마나 많은 아이템을 봐야 하는지 사용자가 간단히 파악할 수 있도록 리스트에 포함된 적합 아이템 수를 알기 쉽게 표시해주는 것도 좋음
- 같은 종류의 아이템을 서비스라도 사용자가 다른 목적을 가지고 있다면 제공해야 할 UI/UX가 다를 수 있으므로 사용자의 어떤 요구에 부응하는 서비스를 만들지 명확히 하는 것이 중요

### 3.2.3 아이템 계열 소비
- 열람, 소비하는 동안 추천된 아이템 계열 전체에서 가치를 누리는 것을 목적으로 하는 경우
  - 예) 음악 스트리밍 서비스
- 아이템 계열 소비 특징은 같은 아이템을 여러 번 추천해도 사용자 만족도가 향상됨
  - 단, 같은 아이템만 추천한다면 새로움이 느껴지지 않아 만족감이 저하되므로 빈도나 시점은 고려해야 함
 
### 3.2.4 서비스 내 체류
- 사용자가 아이템을 열람하는 것 자체를 목적으로 서비스 안을 돌아다니는 경우
  - 구입 의사가 별로 없음에도 과도하게 강요하는 식으로 추천하면 사용자 만족도가 떨어져 재방문하지 않을 가능성이 있음
- 사이트 탐색을 통해 사용자 만족도를 향상시켜두면 구매시 서비스를 이용할 것이라고 기대할 수 있음

## 3.3 서비스 제공자의 목적에 맞는 UI/UX 사례
- 앞에서는 시스템을 사용하는 사용자의 목적에 초점을 맞춰 분류
- 서비스 제공자 측의 대표적인 비즈니스 목적을 5가지로 분류했음 (벤 샤퍼, 스웨어린젠 글 참고)
  - 신규/사용 빈도가 낮은 사용자의 정착
  - 서비스에 대한 신뢰성 향상
  - 사용 빈도 향상 및 이탈 사용자의 복귀
  - 교차 판매(cross selling)
  - 장기적으로 사용자의 충성도 향상

### 3.3.1 신규/사용 빈도가 낮은 사용자의 정착
- 서비스를 처음 방문한 신규 사용자나 사용 빈도가 별로 높지 않은 사용자가 다른 서비스로 이동하기 전에 만족도를 높여 서비스에 정착하도록 하는 것은 서비스 제공자에게 매우 중요
  - 정착을 위해서는 좋은 경험을 제공하고 나쁜 경험을 하지 않도록 해야함
  - **개요 추천(broad recommendation)을 사용**
- 개요 추천 : 서비스 내 데이터 통계 정보나 서비스 지식이 풍부한 편집자의 선택을 기반으로 추천
- 통계 정보에 기반한 추천 :
  - 이번 주 시청 순위
  - 만화 매출 순위 등

### 3.3.2 서비스에 대한 신뢰성 향상
- 사용자는 서비스에서 제공하는 아이템이나 광고가 제공자의 이익을 우선한다고 생각하는 경향이 많음
- 사용자가 제품을 믿고 서비스를 사용하도록 신뢰성을 높여야 함 -> 이런 목적을 달성하는데 **사용자 평가**가 효과적인 경우가 많음
  - 리뷰나 평점을 제공하는 것도 의미가 있음 (다른 사용자의 추천)
  - 신뢰도를 높이면 이탈율을 줄일 수 있음
  - 하지만 어뷰징, 공격적인 문장, 아이템과 관련없는 리뷰는 서비스 신뢰를 저하시킴
  - 아마존의 예시) 게시한 리뷰가 작성된 지역을 표시
  - 리뷰 추천도 좋은 방법
- 사용자가 서비스를 사용할 경우 원하는 바를 달성할 수 있다고 생각하게 하려면 가급적 빠르고 정확하게 사용자 기호에 맞는 아이템을 추천하는 것이 효과적
  - 사용자에게 딱 맞는 아이템을 단번에 추천하는 것은 어려울 것 -> 서비스 안에 자신의 기호에 맞는 아이템이 있을지도 모른다는 기대를 심어주는 것이 중요
  - 넷플릭스의 경우 계정 등록 후 명시적으로 자신이 좋아하는 작품을 선택할 수 있게 함
  - 입력된 정보를 바탕으로 콘텐츠를 개인화하고 있음을 사용자에게 명시적으로 전달

### 3.3.3 사용 빈도 향상 및 이탈 사용자의 복귀
- 이탈한 사용자에게는 **알림 서비스(notification_services)**가 효과적
  - 과거 구매 이력에 기반한 추천 or 사용자가 미리 설정한 조건이 발매되었을 떄 안내하도록 하는 것
- 최근 음성이나 동영상 라이브를 제공하는 서비스가 늘어남 -> 실시간 알림 서비스로 콘텐츠 추천하는 상황이 늘어남
- 알람 서비스는 서비스와 관련된 강한 의지를 보내는 표현이므로 서비스 미사용자들에게 불쾌감을 줄 수 있으니 주의해서 설계
  - 알람 보내는 시점도 고려해야 함

### 3.3.4 교차 판매
- 상품 구매를 검토하고 있는 사용자에게 다른 상품을 함께 또는 개별 구매하도록 추천하는 방법을 **교차 판매(cross selling)**라고 함
- 교차 판매를 구현하는 효과적인 수단은 **연관 아이템 추천(product-assciated recommendations)**를 들 수 있음
  - 사용자가 현재 주목하고 있는 아이템과 연관된 아이템을 사용자에게 표시하는 것
  - 함께 구입하면 얻을 수 있는 이익(할인, 캐시백 등)을 적용시켜 강조하는 것도 효과적
- 사용자가 주목하고있는 아이템과 동등한 유사 아이템을 표시함으로써 상품 비교를 통해 구매하는 것을 도울 수도 있음

### 3.3.5 장기적으로 사용자의 충성도 향상
- 구독 모델(subscription model) 상품 등에서는 장기적으로 사용자 충성도(loyalty)를 높여 서비스를 계속 사용하도록 하는 것이 중요
- 개인화에서는 서비스를 이용할수록 추천에 활용할 수 있는 정보가 축적되어 사용자 만족도가 높아지고 다른 서비스와의 차별로 이어져 장기적으로 사용자 충성도를 향상시킬 수 있음
- 추천은 명시적으로 고객에게 메세지로 전달하는 것이 좋음 (예 : top picks for you)
  - 명시적으로 추천 시스템을 인식시킴에도 실제 추천받은 경험이 좋지 않으면 만족도 악화로 이어짐
  - 반대로 과도한 개인화는 사용자 만족도를 손상시킬 수 있음 (서비스에 내 정보를 빼앗기고 있다는 불안, 추천 기술을 이해할 수 없을때 나오는 불쾌감)

## 3.5 연관 주제
- 추천 시스템의 UI/UX와 깊이 연관된 주제를 몇 가지 소개

### 3.4.1 아이템의 유사도
- 아이템을 추천할 때 유사도가 높기만한 아이템을 추천하는 것이 아니라 아이템 사이의 관계셩을 고려하면 좋음
  - 프린터를 구매한 사용자에게 또 프린터를 추천하면 안될 것. 이 경우 잉크, 토너 등 프린터 부속품과 같은 보완 아이템을 추천하면 좋음
  - 검은색 펜 구매자에게 빨간색 펜 추천은 좋음
- 유사도의 정의는 여러가지로 생각할 수 있음
  - 영화 추천 서비스의 경우 장르, 감독, 출연 배우, 개봉 연도 등에 따라 유사도의 기준이 달라짐

### 3.4.2 신선함, 흥미로움, 다양성
- 신선함(novelty) : 사용자가 관심있는 카테고리 중 사용자가 모르는 아이템을 추천
- 흥미로움(serendipity) : 앞에 나온 신선함에 의외성이라는 요소가 더해진 개념. 사용자가 좋아하는 작가와 유사한 분위기의 신인 작가의 작품을 추천할 경우
- 다양성(diversity) : 위에서 설명한 의외성이라는 감정적인 요소를 정량적으로 측정하기 어려움 -> 다양성으로 정량적으로 측정 가능

- 서비스를 막 시작한 사용자에게는 적합한 아이템을 추천하여 신뢰나 애착을 높이고, 어느 정도 서비스를 계속 사용하는 사용자에게는 흥미로움을 느낄 수 있는 아이템을 추천하여 충성도를 높이는 전략이 좋음

### 3.4.3 추천 아이템 선별
- 추천 아이템 중 평가가 높아도 사용자에게 제시해서는 안되는 아이템도 있음
  - 책의 경우 책을 다시 구매하는 경우는 없음. 하지만 물의 경우는 재구입을 하는 경우가 많음
- 구입으로 연결되지 않는 아이템을 추천 목록에 계속 넣는 것은 사용자 만족도를 저하시킴

### 3.4.4 추천 이유 제시
- **추천 이유**를 제시하면 추천 효과, 투명성, 사용자 만족도를 향상 시킬 수 있음
  - 예시 :
    - 아마존 : 이 상품을 구입한 사람은 이런 상품도 구입했습니다.
    - 넷플릭스 : 평단의 찬사를 받은 드라마
    - 스포티파이 : 수면 카테고리 -> 잠자기 전에 들었던 노래 추천
