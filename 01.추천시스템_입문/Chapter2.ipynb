# CH 2 추천 시스템 프로젝트

## 2.1 추천 시스템 개발에 필요한 3가지 스킬
![데이터 과학자 협회가 공개한 데이터 과학자에게 필요한 스킬셋](./images/img_2-1.png)

- 데이터 과학자 3가지 스킬셋
  - 비즈니스 문제 해결 능력
  - 데이터 사이언스 능력
  - 데이터 엔지니어링 능력

#### 비즈니스 문제 해결 능력
- 비즈니스 관점에서 추천 시스템을 도입함으로써 무엇을 기대할 수 있는지 정의하는 것이 중요
  - 추천 시스템이 필요하지 않거나 인기순 추천만으로도 충분한 경우도 있음
  - 추천 시스템 도입으로 사용자의 어떤 행동 변화를 기대하는지에 대한 KGI(Key Goal Indicator), KPI(Key Performance Indicator) 책정이 매우 중요
  - 사용자의 어떤 행동을 최대하하고 싶은지에 따라 구현할 추천시스템이 달라짐
    - 예) 클릭률, 구입률, 체류율, 트위터 게시 수 등
    - 유튜브는 사용자 전체 시청 시간이 KPI이고 이를 최대화하도록 서비스가 설계되어 있음
      - 기존 시청 동영상 수로 진행했지만 썸네일 이미지가 매력적인 동영상들만 추천하여 변경함
    - KPI 책정 시 현재 얻은 데이터에 필요한 정보가 없다면 지표를 검토하고 해당 데이터가 없다면 개발자와 협력하면서 로그를 정비해야 함

#### 데이터 사이언스 능력
- 추천 시스템 실제 구현에 필요한 능력
  - 다루는 데이터의 양과 종류
  - 비스니스에 필요한 추천 시스템의 구현 복잡도
- 전통적인 방법을 사용해 나오는 정확도를 확인하고 비즈니스 관련 부서와 커뮤니케이션하며 개선하는 것이 중요
  - 그 과정에서 데이터의 성질이나 추천 알고리즘 장단점을 파악
  - 범용적인 추천 알고리즘에서 아이템이나 사용자 정보를 출분히 고려할 수 없는 경우 독자적으로 정식화한 추천 알고리즘을 검토하고 개발해야 함

#### 데이터 엔지니어 능력
- 추천 시스템을 실제 서비스에 조합할 때는 다양한 비즈니스 요건을 고려해 구현하고 시스템을 안정적으로 가동시켜야 함
  - 하루 한 번 사용자에게 추천 상품을 추천할 경우 모든 추천 리스트 작성 연산이 24시간 안에 종료되어야 함

- 이 3가지 스킬을 가진 사람은 많지 않으며 대부분 기업에서 그룹으로 추천 시스템을 구축

## 2.2 추천 시스템 프로젝트 진행 방법
- 추천 프로젝트를 7단계로 나눈 후 전자상거래 사이트를 개선하는 것으로 예시를 들어 설명
  1. 과제 정의
  2. 가설 수립
  3. 데이터 설계/수집/가공
  4. 알고리즘 선정
  5. 학습/파라미터 튜닝
  6. 시스템 구현
  7. 평가 및 개선
  - 1, 2, 7은 비지니스 부서에서, 나머지는 데이터 과학자나 데이터 개발자가 맡아 프로젝트 진행

### 2.2.1 과제 정의
- 비즈니스 상 명확한 목적이 있어야 함 (매출 2배 인상, 유저 2배 인상 등)
  - 이를 비즈니스 지표(KPI)로 결정
  - 1인당 매출을 2배로 늘리기 위해 어디를 늘릴 수 있는지 분석
    - 분석에는 정량적 방법(데이터로 분석), 정성적 분석(인터뷰 등)이 있음
      - 정량적 분석 : 로그 데이터 분석 (검색 사용자 여부, 이탈 페이지 등을 분석)
      - 정성적 분석 : 인터뷰 (잠재적 심리 상태에 관한 인사이트)
      - 이를 바탕으로 과제 중요도를 결정 (아래의 표는 예시)
        ![과제와 중요도](./images/tbl_2-1.png)
    - 데이터 분석의 결과로 아래의 두 가지 과제가 중요한 것으로 판단
      - 너무 많은 아이템이 표시되어 원하는 아이템에 도달하지 못해 이탈률이 높음
      - 각 아이템의 상세 화면에서 해당 아이템과 비슷한 아이템을 알 수 없어 이탈률이 높음

### 2.2.2 가설 수립
- 각 과제 해결 방법과 그것을 실현하기 위한 비용을 검토하고 비용 대비 효과(ROI)가 높은 목표 달성을 위한 구체적인 목표부터 실행
  ![각 대응책의 우선 순위](./images/tbl_2-2.png)

- 위 경우에는 먼저 검색과 관련된 구현 비용이 낮고 중요도가 높은 과제를 해결한 뒤 추천 시스템을 다루는 것이 효과적
- 요구사항을 잘 들어보며 검색 시스템 개선이나 디자인 변경으로 CTR이나 전환율을 높인 사례가 있음
- 비즈니스 단계에서 적절한 추천 아이템을 선정

### 2.2.3 데이터 설계/수집/가공
- 현재 어떤 데이터가 쌓이고 있는지 정리
  - 사용자와 아이템의 콘텐츠 정보
  - 사용자와 아이템의 인터랙션 정보
- 자사 서비스가 가지고 있는 데이터로 추천 시스템을 구축할 수 있는지 검토
- 태그 정보등의 데이터가 있다면 데이터가 의미에 맞게 잘 들어갔는지 확인
- 필요한 데이터가 적절하게 존재하는지 확인 후 추천 시스템을 사용할 수 있도록 전처리

### 2.2.4 알고리즘 선정
- 비즈니스 목표에 적합한 알고리즘 선정
  - 알고리즘 계산 시간
  - 필요한 데이터
  - 예측 정확도등
- 정확도를 올릴수록 비용이 급격하게 증가할 수 있으므로 서비스 초기에는 간단한 알고리즘부터 구현(아래 그림 참고)
  ![비용 대비 효과의 관계](./images/img_2-2.png)

### 2.2.5 학습/파라미터 튜닝
- 학습, 튜닝을 수행하고 과거 데이터로 알고리즘을 검증
- Recall, Precision 지표 확인 뿐만 아니라 결과 아이템을 확인하고 적합한지 판단
  - 타 부서와 커뮤니케이션할 때는 결과 아이템으로 구체적인 사례를 제시하면 좋음
- 온라인에서 검증할 때는 편향(bias)에 주의
  - 검색에서는 상위에 있는 아이템이 선택받기 쉬움
  - 비디오 평가 데이터는 5점이 많다는 편향이 있음

### 2.2.6 시스템 구현
- 시스템 구현에서 고려할 만한 것들
  - 추천 알고리즘의 학습이나 예측 변경 빈도
  - 신규 아이템이나 사용자에 대한 추천
  - 추천에 연관된 데이터 파이프라인 설계
- 배치 추천과 실시간 추천에 관해 간단히 설명(자세한 내용은 6장)
  #### 배치 추천
  - 정해진 시점에 추천 목록을 업데이트하여 사용자에게 제공하는 것
    - 추천 목록 : 추천 모델 학습, 추천 모델에 의한 예측값의 2단계 처리를 통해 작성됨
      - 추천 모델 학습 : 신규 아이템 추천 (비즈니스 목적에 따라 배치 주기 결정)
      - 추천 모델에 의한 예측 : 사용자가 많아질수록 시간이 늘어나므로 병렬처리를 하는 등 시간을 줄이는 방법을 고안
  #### 실시간 추천
  - 사용자 행동 이력을 즉시 반영해 추천 목록을 만드는 것
  - 데이터의 동기나 실시간 연산 등 폭넓은 엔지니어링 기술이 필요함

### 2.2.7 평가 및 개선
- 추천 기능을 출시하고 효과가 있는지 검증
- A/B 테스트 같은 구조로 평가하면 좋음
- 추천 시스템 도입 후 발생한 매출이라도 추천 때문인지 확인 필요
- 시스템 전체를 확인하고 악영향이 없는지 확인하는 것도 중요
