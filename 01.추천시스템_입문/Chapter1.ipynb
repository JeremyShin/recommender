# CH 1 추천 시스템
- 추천 시스템의 개요와 역할, 사용자에게 어떤 가치를 전달할지에 대한 설명

## 1.1 추천 시스템
- 유저는 손쉽게 마음에 드는 아이템을 만나고 기업은 더 많은 상품을 팔거나 긴 시간 서비스에 머무르게 할 수 있음
- 여러 후보 가운데 가치 있는 것을 선정해서 의사 결정을 지원하는 시스템
  - 가치를 정의하는 방법에 따라 다양한 추천 알고리즘이 제안됨
- 추천 시스템 정의에서 '의사 결정을 지원한다'는 것은 선택한 아이템을 실제 사용자가 열람 또는 구입하도록 제시하는 것이 중요
  - 아이템은 어떤 단계로, 어떻게 사용자에게 도달하도록 할지 설계하는 것이 중요
    - 예시
      - 방법1 : 웹 사이트에서 추천 아이템을 표시
      - 방법2 : 메일
      - 방법3 : 스마트폰 푸시 알림
      - 시점1 : 상품 구입 시점에 다른 아이템 추천
      - 시점2 : 상품을 살펴보는 중에 추천 

## 1.2 추천 시스템의 역사
- 제록스 팔로알토 연구소의 연구자 골드버그가 협조 필터링 제안
- 제안한 시스템은 태피스트리 : 늘어나는 전자 메일 중 유익한 메일을 선택하는 것
- 아마존에서 원하느 상품을 제공하는 기술로 추천 시스템 활용

## 1.3 추천 시스템의 종류
- 추천시스템은 입력(데이터 입력), 프로세스(추천 설계), 출력(추천 결과 제시)의 3가지 요소로 정리
![추천시스템3요소](./images/1-3.png)

### 1.3.1 입력(데이터 입력)
- 데이터는 크게 두 종류가 있음
  - 콘텐츠(contents)
  - 인터렉션(interaction)
#### 사용자와 아이템의 콘텐츠 정보
- 사용자의 콘텐츠 정보 : 프로필 정보 (나이, 성별, 주소, 선호 카테고리나 가격대 등)
- 아이템의 콘텐츠 정보 : 카테고리, 상품 설명문, 발메일, 가격, 제작자 등
- **내용 기반 필터링** : 콘텐츠 정보를 이용하는 추천

#### 사용자와 아이템의 인터렉션 정보
- 사용자가 서비스 안에서 행동한 이력 데이터 (열람, 구입, 북마크, 평가 등)
- **협조 필터링** : 인터랙션 정보를 사용하는 추천
- 신규 사용자나 신규 아이템은 행동 이력 데이터가 없으므로 전자의 콘텐츠 정보를 사용한 내용 기반 필터링으로 추천하는 경우가 많음
- **콜드 스타트 문제** : 신규 사용자나 신규 아이템 데이터가 적어 추천이 어려워지는 것

### 1.3.2 프로세스 (추천 설계)
- 입력 데이터를 활용해 추천하는 방법은 3가지

#### 개요 추천(개인화 없음)
- 모든 사용자에게 동일한 내용을 제시(신규 아이템순, 낮은 가격순 등)
- 경우에 따라 개인화 적용한 것보다 클릭률, 구매율이 높을수도 있음
  - 미디어 -> 정보의 신규성이 중요
  - 모두가 보고 있는 정보를 보고 싶다 -> 인기순

#### 연관 아이템 추천
- 아이템 페이지 아래쪽에 표시되는 경우가 많음
- 각 아이템 사이의 유사도를 사용
  - 유사도 설계는 내용 기반 필터링, 협조 필터링(협업 필터링) 방법이 있음
- 일반적으로 업조 필터링 방법이 카테고리, 키워드로는 표현할 수 없는 아이템 분위기나 콘셉을 유사도에 반영한 경우가 많음
- 유사도 판단 시 고려할 점
  - 비슷한 아이템이 가능성이 높을 때
    - 프린터 구입 시 비슷한 종류를 많이 확인
  - 함꼐 구입될 만한 아이템이 가능성이 높을 때
    - 프린터 구입시 카트리지와 같은 소모품을 함께 제시
- **해리 포터 문제**
  - 특정 시기에 많은 사람이 해리 포터 서적을 다른 아이템과 함꼐 구매

#### 개인화 추천
- 사용자 프로필이나 행동 이력을 기반으로 각 사용자에 맞춰 추천하는 것
- 개인화 추천의 방법
  - 콘텐츠 기반
  - 인터랙션 데이터 활용
  - 위의 두 가지 방법의 혼합
- 콘텐츠 기반
  - 나이, 거주지 등의 프로필 정보를 기반으로 그에 맞는 아이템 그룹 설계
- 인터렉션 데이터 기반
  - 협업 필터링 등의 방법으로 사용자 과거 행동 이력으로부터 추천 아이템 계산
  - 열람 이력을 그대로 표시하는 추천은 구현 비용이 적은 것에 비해 효과가 높음
- 열람 이력과 연관 아이템 추천 구조를 조합해서 사용
  - 사용자가 마지막 열람한 아이템과 비슷한 아이템 그룹을 추출해 메일 송부

### 1.3.3 출력(추천 결과 제시)
- 좋은 추천 아이템을 설계해도 제시 방법이 나쁘면 사용자의 행동을 이끌어 낼 수 없음
- 자세한 제시 방법은 3장에서 소개
- 추천 시스템을 도입할 때 이 틀에서 생각해보면 좋음
  - 어떤 데이터가 있는지
  - 어떤 알고리즘으로 비즈니스에 영향을 줄지
  - 어떤 제시 방법을 사용해야 사용자가 행동할지

## 1.4 검색 시스템과 추천 시스템
- 추천시스템과 비슷한 검색 시스템이 있음
- 역할의 차이를 이해하자

### 1.4.1 검색 시스템
- 검색 시스템 : 키워드를 입력해서 원하는 문장을 찾아내는 기술
- 검색 시스템은 구글 창업자가 발명했으며 페이지 랭크 알고리즘이 많이 알려져 있음

### 1.4.2 검색 시스템과 추천 시스템 비교
- 공통점 : 많은 아이템에서 가치 있는 아이템을 선택
- 사용자는 이 둘을 다른 목적으로 사용 (push, pull 타입으로 부르기도 함)
![검색,추천시스템비교](./images/1-1_table.png)

- 검색 : 능동적 (pull 타입)
  - 최근에는 검색 결과를 개인화하는 경우도 증가하고 있음
- 추천 : 원하는 것 파악하지 않더라도 사용자가 좋아하는 상품을 제시(push 타입)
  - 사용자가 몰랐던 아이템을 만나기도 함
  - 사용자가 좋아하는 음악, 패션의 분위기 등을 먼저 찾아주기도 함
- 검색과 추천 사용 비중은 서비스에 따라 다름
- 추천 시스템은 사용자 과거 행동 이력에서 기호를 추출하는 기술을 발전시켜 옴
- 사용자의 의도를 더 정확하게 추측하려는 연구가 증가
